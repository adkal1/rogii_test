```python
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
from IPython.display import display, clear_output
import ipywidgets as widgets
```

1. Чтение csv-файлов

```python
dataset1 = pd.read_csv(open('datas/dataset_1.csv', 'r', encoding='cp1251'), sep=',', low_memory=False)
dataset2 = pd.read_csv(open('datas/dataset_2.csv', 'r', encoding='cp1251'), sep=',', low_memory=False)
dataset3 = pd.read_csv(open('datas/dataset_3.csv', 'r', encoding='cp1251'), sep=',', low_memory=False)
```

2. Удаление единиц измерений из таблиц

```python
dataset1.drop(labels=[0], axis =0, inplace=True)
dataset2.drop(labels=[0], axis =0, inplace=True)
dataset3.drop(labels=[0], axis =0, inplace=True)
```

3. Объедение таблиц

```python
datasets = pd.concat([dataset1, dataset2, dataset3], axis=0)
datasets.drop(['Unnamed: 0', 'Time'], axis=1)
```

4. Изменение названия первого столбца на DataTime

```python
datasets.rename(columns={'RIGTIME': 'DateTime'}, inplace=True)
```

5. Изменение формата столбца "DateTime" в "YYY-mm-dd HH:MM:SS"

```python
datasets['DateTime'] = pd.to_datetime(datasets['DateTime'], format="%Y-%m-%d %H:%M:%S" )
```

```python
datasets = datasets.sort_values(by='DateTime')
```

6. Приведение столбцов в формат float

```python
datasets[['HOLEDEPTH', 'RPMA', 'BLKPOS', 'ROPA', 'HKLA', 'PRES', 'DEPTBITM', 'SPM1', 'SPM2', 'SPM3', 'MFIA', 'GAS', 'TORQX', 'DIFP', 'WOBX']]\
= datasets[['HOLEDEPTH', 'RPMA', 'BLKPOS', 'ROPA', 'HKLA', 'PRES', 'DEPTBITM', 'SPM1', 'SPM2', 'SPM3', 'MFIA', 'GAS', 'TORQX', 'DIFP', 'WOBX']]\
.astype(float)
```

7. Замена пропущенных данных средними значениями для столбца "SPM2", медианными значениями для столбца "SPM3"

```python
datasets['SPM2'] = datasets['SPM2'].fillna(datasets['SPM2'].mean())
datasets['SPM3'] = datasets['SPM3'].fillna(datasets['SPM3'].median())
```

8. Создание GUI для выбора графика по различным параметрам

```python
def update_plot(change):
    clear_output(wait=True)
    for column in datasets.columns[1:16]:
        if change['new'] == column:
            display(datasets.plot(x="DateTime", y=column))
plot_selector = widgets.ToggleButtons(options=datasets.columns[1:16], description='Plot:')
plot_selector.observe(update_plot, names='value')
display(plot_selector)
```

9. Создание нового DataFrame, в котором данные будут распределены по столбцу HOLEDEPTH.

```python
datasets_new = datasets.reindex(columns=['HOLEDEPTH', 'DateTime', 'RPMA', 'BLKPOS', 'ROPA', 'HKLA', 'PRES', 'DEPTBITM', 'SPM1', 'SPM2', 'SPM3', 'MFIA', 'GAS', 'TORQX', 'DIFP', 'WOBX'])
```
